```python
from dotenv import load_dotenv

load_dotenv()
```

```python
# 음성 입력
import speech_recognition as sr

recognizer = sr.Recognizer()
texts = []

while True:
    with sr.Microphone() as source:
        print('''
              희망하는 요구사항을 말씀해주세요~
              (예시: 바다 근처가 좋아, 문화 유산이 많으면 좋겠어, 사람이 적은 곳이 좋아 등..) 
              ''')
        audio = recognizer.listen(source)
        text = recognizer.recognize_google(audio, language='ko-KR')
        print(text)

        if text == '종료':
            break

        texts.append(text)

needs = ' '.join(texts)
print(needs)
```

```python
# 테스트

needs = '나는 바다가 좋고, 사람이 적고, 수도권에 가까웠으면 좋겠어.'
```

```python
# LLM 구현
from openai import OpenAI

client = OpenAI()

prompt = f'''
    사용자 여행 조건 : {needs}

이 조건에 맞는 국내 여행지 3곳을 추천해주고,
각 추천 이유를 2줄로 나타내주고,
그리고 본인이 판단한 평점을 10점 만점으로 평가해줘.

추가적으로, 주의 사항으로는 Markdown 기호는 절대 쓰지마.
형식은
추천 여행지 세 곳을 안내해드릴게요.

첫 번째 추천지는 [장소명]입니다.
추천 이유 첫 번째, [이유]
추천 이유 두 번째, [이유]

두 번째 추천지는 [장소명]입니다.
추천 이유 첫 번째, [이유]
추천 이유 두 번째, [이유]

세 번째 추천지는 [장소명]입니다.
추천 이유 첫 번째, [이유]
추천 이유 두 번째, [이유]

마무리는 총 정리해서 작성해줘
'''

response = client.responses.create(
    model="gpt-4.1-mini",
    input=prompt
)

answer = response.output_text
print(answer)
```

```python
# TTS 변환
from gtts import gTTS
from IPython.display import Audio

tts_output = gTTS(answer, lang='ko')
tts_output.save('gtts_output.mp3')

display(Audio('gtts_output.mp3', autoplay=True))
```
